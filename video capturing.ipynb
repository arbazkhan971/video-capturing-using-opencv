```python
import cv2
import time
```

```python
video = cv2.VideoCapture(0)
a=0

while True:
   a+=1
   check , frame = video.read()
   print(check)
   print(frame)
   cv2.imshow("captions",frame)
   key=cv2.waitKey(1)
   if key == ord('q'):
        break
        

video.release()
cv2.destroyAllWindows()
```
